```python
import pandas as pd
import requests
import time

# Function to fetch repositories
def fetch_repos(user_login, token):
    url = f'https://api.github.com/users/{user_login}/repos'
    headers = {'Authorization': f'token {token}'}
    response = requests.get(url, headers=headers)
    return response.json()

# Function to process repositories
def process_repos(repos, user_login):
    repo_data = []
    for repo in repos[:500]:  # Limit to 500 most recently pushed repositories
        repo_info = {
            'login': user_login,
            'full_name': repo['full_name'],
            'created_at': repo['created_at'],
            'stargazers_count': repo['stargazers_count'],
            'watchers_count': repo['watchers_count'],
            'language': repo['language'],
            'has_projects': repo['has_projects'],
            'has_wiki': repo['has_wiki'],
            'license_name': repo['license']['name'] if repo['license'] else None
        }
        repo_data.append(repo_info)
    return pd.DataFrame(repo_data)

# Read users.csv
users_df = pd.read_csv('users.csv')

# Personal access token
api_key_file = 'D:/Tds_work/token.txt'

if os.path.exists(api_key_file):
    with open(api_key_file, 'r') as file:
        token = file.read().strip()

# Fetch and process repositories
all_repos_df = pd.DataFrame()
for index, row in users_df.iterrows():
    user_login = row['login']
    repos = fetch_repos(user_login, token)
    repos_df = process_repos(repos, user_login)
    all_repos_df = pd.concat([all_repos_df, repos_df], ignore_index=True)
    time.sleep(2)  # Add delay to respect rate limits

# Save to CSV
all_repos_df.to_csv('repositories.csv', index=False)
print("CSV file created: repositories.csv")
```

```python
import os
import asyncio
import aiohttp
import aiofiles
import pandas as pd
import nest_asyncio

nest_asyncio.apply()  # Allow nested event loops

async def fetch_repos(session, user_login, token):
    url = f'https://api.github.com/users/{user_login}/repos'
    headers = {'Authorization': f'token {token}'}
    async with session.get(url, headers=headers) as response:
        return await response.json()

async def process_repos(session, user_login, token):
    repos = await fetch_repos(session, user_login, token)
    repo_data = []
    for repo in repos[:500]:  # Limit to 500 most recently pushed repositories
        repo_info = {
            'login': user_login,
            'full_name': repo['full_name'],
            'created_at': repo['created_at'],
            'stargazers_count': repo['stargazers_count'],
            'watchers_count': repo['watchers_count'],
            'language': repo['language'],
            'has_projects': repo['has_projects'],
            'has_wiki': repo['has_wiki'],
            'license_name': repo['license']['name'] if repo['license'] else None
        }
        repo_data.append(repo_info)
    return pd.DataFrame(repo_data)

async def main():
    token_file = 'D:/Tds_work/token.txt'
    async with aiofiles.open(token_file, 'r') as file:
        token = await file.read()
        token = token.strip()  # Ensure no extra spaces or newlines

    users_df = pd.read_csv('users.csv')
    all_repos_df = pd.DataFrame()

    async with aiohttp.ClientSession() as session:
        tasks = []
        for index, row in users_df.iterrows():
            user_login = row['login']
            tasks.append(process_repos(session, user_login, token))

        repos_dataframes = await asyncio.gather(*tasks)
        all_repos_df = pd.concat(repos_dataframes, ignore_index=True)

    all_repos_df.to_csv('repositories.csv', index=False)
    print("CSV file created: repositories.csv")

asyncio.run(main())
```

```python
import pandas as pd

# Read the existing CSV
df = pd.read_csv('users.csv')

# Convert 'created_at' to datetime
df['created_at'] = pd.to_datetime(df['created_at'])

# Filter for weekend commits
df['day_of_week'] = df['created_at'].dt.dayofweek
weekend_commits = df[df['day_of_week'] >= 4]
top_users = weekend_commits.groupby('login')['public_repos'].sum().sort_values(ascending=False).head(5)
top_users = weekend_commits.groupby('login')['public_repos'].sum().sort_values(ascending=False).head(5)
top_user_logins = ', '.join(top_users.index.tolist())
print(f"Top 5 users who created the most repositories on weekends (UTC): {top_user_logins}")
```

```python
import pandas as pd

# Read the existing CSV
df = pd.read_csv('users.csv')

# Ensure the 'following' column is numeric
df['following'] = pd.to_numeric(df['following'], errors='coerce')

# Filter the data
hireable_users = df[df['hireable'] == 'true']
non_hireable_users = df[df['hireable'] != 'true']

# Calculate the average number of people followed by each group
avg_following_hireable = hireable_users['following'].mean()
avg_following_non_hireable = non_hireable_users['following'].mean()

# Calculate the difference
difference = avg_following_hireable - avg_following_non_hireable

print(f"The average following difference is {difference:.3f}")
```
